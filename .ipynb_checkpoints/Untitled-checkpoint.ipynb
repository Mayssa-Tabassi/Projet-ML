```python
df = df.dropna(subset=['Categorie'])

print("\nAprès suppression des catégories manquantes :")
print(df.shape)
```

```python
if 'Source' in df.columns:
    df = df.drop(columns=['Source'],axis=1)

print("\nColonnes restantes :")
print(df.columns)
```

```python
df.head()
```

```python
# -----------------------------
# 4. Imputation par catégorie
# -----------------------------
numeric_cols = [
    'Poids',
    'Volume',
    'Conductivite',
    'Opacite',
    'Rigidite',
    'Prix_Revente'
]
for col in numeric_cols:
    df[col] = df.groupby('Categorie')[col].transform(
        lambda x: x.fillna(x.median())
    )

# Si une catégorie entière est vide
for col in numeric_cols:
    df[col] = df[col].fillna(df[col].median())
```

```python
for col in numeric_cols:
    before = df[col].copy()

    corrected = df.groupby('Categorie')[col].transform(cap_outliers)

    nb = (before != corrected).sum()

    print(f"{col} : {nb} valeurs corrigées")

    df[col] = corrected
```

```python
# -----------------------------
# Traitement des valeurs aberrantes par catégorie
# -----------------------------
for col in numeric_cols:

    def cap_outliers(x):
        Q1 = x.quantile(0.25)
        Q3 = x.quantile(0.75)
        IQR = Q3 - Q1

        borne_inf = Q1 - 1.5 * IQR
        borne_sup = Q3 + 1.5 * IQR

        return x.clip(borne_inf, borne_sup)

    df[col] = df.groupby('Categorie')[col].transform(cap_outliers)

print("Traitement terminé")
print(df.head())
```
