## 📂 1. Chargement du Dataset

````python
DATA_PATH     = 'outputs/full_cleaned.csv'
ORIG_CSV_PATH = 'dataset_ProjetML_2026 (1).csv'
TEXT_COL      = 'Rapport_Collecte'
LABEL_COL     = 'Categorie'

if not os.path.exists(DATA_PATH):
    raise FileNotFoundError(
        f'Fichier introuvable : {DATA_PATH}\n'
        'Exécuter Module 1 en premier pour générer full_cleaned.csv'
    )

df = pd.read_csv(DATA_PATH)
print(f'✅ full_cleaned.csv chargé  →  {df.shape}')

# Récupération de Rapport_Collecte si absente
if TEXT_COL not in df.columns:
    if not os.path.exists(ORIG_CSV_PATH):
        raise FileNotFoundError(
            f'CSV original introuvable : {ORIG_CSV_PATH}\n'
            f'Colonnes disponibles : {list(df.columns)}'
        )
    df_orig = pd.read_csv(ORIG_CSV_PATH)
    df_orig_aligned = df_orig.dropna(subset=[LABEL_COL]).reset_index(drop=True)
    df[TEXT_COL] = df_orig_aligned[TEXT_COL].values[:len(df)]
    print(f'✅ {TEXT_COL} mergée depuis {ORIG_CSV_PATH}')
else:
    print(f'✅ {TEXT_COL} déjà présente')

n_avant = len(df)
df = df[[TEXT_COL, LABEL_COL]].dropna().reset_index(drop=True)
print(f'   Lignes retenues : {len(df)} / {n_avant}')
print(f'   Distribution des classes :')
print(df[LABEL_COL].value_counts())
df.head(3)
````

## 🔤 2. Prétraitement Textuel

````python
# Stopwords : français NLTK + domaine métier
STOP_FR = set(stopwords.words('french'))

STOP_DOMAIN = {
    'utilisation', 'usage', 'haute', 'haut', 'bon', 'bonne',
    'grand', 'grande', 'type', 'produit', 'matériau', 'matériaux',
    'excellent', 'excellente', 'faible', 'élevé', 'élevée',
    'moyen', 'moyenne', 'application', 'applications',
}

ALL_STOPS = STOP_FR | STOP_DOMAIN

# Stemmer SnowballStemmer (français)
stemmer = SnowballStemmer('french')


def clean_text(text):
    """Nettoyage : minuscules + suppression caractères spéciaux."""
    text = str(text).lower()
    text = re.sub(r'[^a-zàâäéèêëîïôùûüÿç\s]', ' ', text)
    text = re.sub(r'\s+', ' ', text).strip()
    return text


def tokenize(text):
    """Tokenisation NLTK French."""
    return word_tokenize(text, language='french')


def remove_stopwords(tokens):
    """Suppression stopwords français + domaine, longueur > 2."""
    return [t for t in tokens if t not in ALL_STOPS and len(t) > 2]


def stem_tokens(tokens):
    """Stemming SnowballStemmer français."""
    return [stemmer.stem(t) for t in tokens]


def lemmatize_tokens_spacy(text):
    """
    CORRECTION : Lemmatisation SpaCy FR (fr_core_news_sm)
    Remplace WordNetLemmatizer qui est un lemmatiseur ANGLAIS —
    inadapté pour du texte français.
    """
    if not SPACY_OK:
        return stem_tokens(tokenize(clean_text(text)))
    doc = nlp_spacy(text)
    tokens = [token.lemma_.lower() for token in doc
              if not token.is_stop and not token.is_punct
              and len(token.lemma_) > 2
              and token.lemma_.lower() not in ALL_STOPS]
    return tokens


def preprocess(text, method='stem'):
    """Pipeline complet : nettoyage → tokenisation → stopwords → stem/lemma."""
    cleaned = clean_text(text)
    if method == 'lemma' and SPACY_OK:
        tokens = lemmatize_tokens_spacy(cleaned)
    else:
        tokens = remove_stopwords(tokenize(cleaned))
        if method == 'stem':
            tokens = stem_tokens(tokens)
    return ' '.join(tokens)


# Application aux deux variantes
df['text_stem']  = df[TEXT_COL].apply(lambda x: preprocess(x, method='stem'))
df['text_lemma'] = df[TEXT_COL].apply(lambda x: preprocess(x, method='lemma'))
df['tokens']     = df[TEXT_COL].apply(lambda x: remove_stopwords(tokenize(clean_text(x))))

print('Exemple prétraitement :')
idx = 0
print(f'  Original : {df[TEXT_COL].iloc[idx]}')
print(f'  Stemming : {df["text_stem"].iloc[idx]}')
print(f'  Lemma FR : {df["text_lemma"].iloc[idx]}')
````

## ✂️ 3. Split Train / Val / Test — 70 / 15 / 15

> **CORRECTION** : Le CDC impose un split **70:15:15** avec stratification.
> La version initiale utilisait un split 80/20 sans ensemble de validation.

````python
le = LabelEncoder()
y  = le.fit_transform(df[LABEL_COL])

X_stem  = df['text_stem'].values
X_lemma = df['text_lemma'].values
tokens  = df['tokens'].tolist()

# ── Étape 1 : séparer le test set (15%) ──────────────────────
X_train_val_s, X_test_s, y_train_val, y_test = train_test_split(
    X_stem, y, test_size=0.15, stratify=y, random_state=SEED
)
X_train_val_l, X_test_l, _, _ = train_test_split(
    X_lemma, y, test_size=0.15, stratify=y, random_state=SEED
)
tokens_train_val, tokens_test, _, _ = train_test_split(
    tokens, y, test_size=0.15, stratify=y, random_state=SEED
)

# ── Étape 2 : séparer val (15% du total = ~17.6% du reste) ──
val_ratio = 0.15 / 0.85  # ~0.1765

X_train_s, X_val_s, y_train, y_val = train_test_split(
    X_train_val_s, y_train_val,
    test_size=val_ratio, stratify=y_train_val, random_state=SEED
)
X_train_l, X_val_l, _, _ = train_test_split(
    X_train_val_l, y_train_val,
    test_size=val_ratio, stratify=y_train_val, random_state=SEED
)
tokens_train, tokens_val, _, _ = train_test_split(
    tokens_train_val, y_train_val,
    test_size=val_ratio, stratify=y_train_val, random_state=SEED
)

total = len(y)
print(f'Split 70/15/15 (total = {total} lignes) :')
print(f'  Train : {len(y_train):4d} ({len(y_train)/total*100:.1f}%)')
print(f'  Val   : {len(y_val):4d} ({len(y_val)/total*100:.1f}%)')
print(f'  Test  : {len(y_test):4d} ({len(y_test)/total*100:.1f}%)')
print(f'  Classes : {list(le.classes_)}')
````

## 📦 4. Vectorisation

````python
# ── 4.1 Bag of Words (Baseline) ──────────────────────────────
bow_vec = CountVectorizer(max_features=5000, ngram_range=(1, 1))
X_bow_train = bow_vec.fit_transform(X_train_s)
X_bow_val   = bow_vec.transform(X_val_s)
X_bow_test  = bow_vec.transform(X_test_s)
print(f'BoW    → train:{X_bow_train.shape}  val:{X_bow_val.shape}  test:{X_bow_test.shape}')

# ── 4.2 TF-IDF uni+bigrammes (Référence) ────────────────────
tfidf_vec = TfidfVectorizer(
    max_features=8000, ngram_range=(1, 2),
    sublinear_tf=True, min_df=2
)
X_tfidf_train = tfidf_vec.fit_transform(X_train_s)
X_tfidf_val   = tfidf_vec.transform(X_val_s)
X_tfidf_test  = tfidf_vec.transform(X_test_s)
print(f'TF-IDF → train:{X_tfidf_train.shape}  val:{X_tfidf_val.shape}  test:{X_tfidf_test.shape}')

# Top termes TF-IDF par classe
feature_names = np.array(tfidf_vec.get_feature_names_out())
print('\nTop 5 termes TF-IDF par classe :')
for cls_idx, cls_name in enumerate(le.classes_):
    mask   = y_train == cls_idx
    scores = X_tfidf_train[mask].mean(axis=0).A1
    top5   = feature_names[np.argsort(scores)[-5:][::-1]]
    print(f'  {cls_name:15s}: {list(top5)}')
````

````python
# ── 4.3 Word2Vec (embeddings sémantiques) ────────────────────
w2v_model = Word2Vec(
    sentences=tokens_train,
    vector_size=100, window=5, min_count=1,
    workers=4, sg=1, seed=SEED, epochs=50
)

def text_to_w2v(tokens_list, model):
    """Moyenne des vecteurs Word2Vec par document."""
    vectors = []
    for toks in tokens_list:
        vecs = [model.wv[t] for t in toks if t in model.wv]
        vectors.append(np.mean(vecs, axis=0) if vecs else np.zeros(model.vector_size))
    return np.array(vectors)

X_w2v_train = text_to_w2v(tokens_train, w2v_model)
X_w2v_val   = text_to_w2v(tokens_val,   w2v_model)
X_w2v_test  = text_to_w2v(tokens_test,  w2v_model)
print(f'Word2Vec → train:{X_w2v_train.shape}  val:{X_w2v_val.shape}  test:{X_w2v_test.shape}')

# ── 4.4 FastText (robustesse OOV) ────────────────────────────
ft_model = GensimFastText(
    sentences=tokens_train,
    vector_size=100, window=5, min_count=1,
    workers=4, sg=1, seed=SEED, epochs=50
)

def text_to_ft(tokens_list, model):
    """Moyenne des vecteurs FastText (gère OOV via sous-mots)."""
    vectors = []
    for toks in tokens_list:
        vecs = [model.wv[t] for t in toks]
        vectors.append(np.mean(vecs, axis=0) if vecs else np.zeros(model.vector_size))
    return np.array(vectors)

X_ft_train = text_to_ft(tokens_train, ft_model)
X_ft_val   = text_to_ft(tokens_val,   ft_model)
X_ft_test  = text_to_ft(tokens_test,  ft_model)
print(f'FastText → train:{X_ft_train.shape}  val:{X_ft_val.shape}  test:{X_ft_test.shape}')

# Test robustesse OOV
oov_word = 'metalique'
vec_oov = ft_model.wv[oov_word]
print(f'\nFastText OOV "{oov_word}" → dim={vec_oov.shape}, non-nul={np.any(vec_oov)}')
````

## 🤖 5. Classification + Logging MLflow

> **CORRECTION** : Chaque expérience est maintenant loguée dans MLflow
> (accuracy, F1, paramètres du modèle). Le CDC exige ≥ 5 expériences.

````python
results = []
CLASSES = list(le.classes_)


def evaluate(name, clf, X_tr, X_te, y_tr, y_te,
             X_val=None, y_val=None,
             vectorizer_name='unknown', log_mlflow=True):
    """
    Entraîne un classificateur, évalue sur val + test,
    et logue l'expérience dans MLflow.
    """
    clf.fit(X_tr, y_tr)

    # Évaluation test
    y_pred_test = clf.predict(X_te)
    acc_test = accuracy_score(y_te, y_pred_test)
    f1_test  = f1_score(y_te, y_pred_test, average='weighted')

    # Évaluation val (si fourni)
    acc_val, f1_val = None, None
    if X_val is not None and y_val is not None:
        y_pred_val = clf.predict(X_val)
        acc_val = accuracy_score(y_val, y_pred_val)
        f1_val  = f1_score(y_val, y_pred_val, average='weighted')

    results.append({
        'Modèle'      : name,
        'Vectorizer'  : vectorizer_name,
        'Acc_Val'     : acc_val,
        'F1_Val'      : f1_val,
        'Acc_Test'    : acc_test,
        'F1_Test'     : f1_test,
    })

    # ── Logging MLflow ────────────────────────────────────────
    if log_mlflow:
        with mlflow.start_run(run_name=name):
            # Paramètres
            mlflow.log_param('vectorizer', vectorizer_name)
            mlflow.log_param('classifier', clf.__class__.__name__)
            clf_params = clf.get_params()
            for k, v in clf_params.items():
                mlflow.log_param(f'clf_{k}', str(v))

            # Métriques
            mlflow.log_metric('accuracy_test', acc_test)
            mlflow.log_metric('f1_weighted_test', f1_test)
            if acc_val is not None:
                mlflow.log_metric('accuracy_val', acc_val)
                mlflow.log_metric('f1_weighted_val', f1_val)

            # Enregistrement du modèle sklearn
            mlflow.sklearn.log_model(clf, artifact_path='model')

    val_str = f'Val_F1={f1_val:.4f}  ' if f1_val is not None else ''
    print(f'{name:45s}  {val_str}Test_Acc={acc_test:.4f}  Test_F1={f1_test:.4f}')
    return clf, y_pred_test


def plot_confusion(y_te, y_pred, classes, title):
    cm = confusion_matrix(y_te, y_pred)
    plt.figure(figsize=(6, 5))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                xticklabels=classes, yticklabels=classes)
    plt.title(title)
    plt.ylabel('Réel')
    plt.xlabel('Prédit')
    plt.tight_layout()
    plt.savefig(f'outputs/{title.replace(" ", "_")}.png', dpi=120)
    plt.show()


print('✅ Fonctions evaluate() + MLflow logging prêtes')
````

````python
print('=== BoW — Baseline ===')
evaluate('BoW + Naive Bayes Multinomial',
         MultinomialNB(),
         X_bow_train, X_bow_test, y_train, y_test,
         X_bow_val, y_val, vectorizer_name='BoW')

evaluate('BoW + Logistic Regression',
         LogisticRegression(max_iter=1000, C=1.0, random_state=SEED),
         X_bow_train, X_bow_test, y_train, y_test,
         X_bow_val, y_val, vectorizer_name='BoW')

evaluate('BoW + LinearSVC',
         LinearSVC(C=1.0, max_iter=2000, random_state=SEED),
         X_bow_train, X_bow_test, y_train, y_test,
         X_bow_val, y_val, vectorizer_name='BoW')

evaluate('BoW + Random Forest',
         RandomForestClassifier(n_estimators=200, random_state=SEED),
         X_bow_train, X_bow_test, y_train, y_test,
         X_bow_val, y_val, vectorizer_name='BoW')
````

````python
print('=== TF-IDF (uni+bigrammes) — Référence ===')

evaluate('TF-IDF + Naive Bayes Complement',
         ComplementNB(),
         X_tfidf_train, X_tfidf_test, y_train, y_test,
         X_tfidf_val, y_val, vectorizer_name='TF-IDF')

evaluate('TF-IDF + Logistic Regression',
         LogisticRegression(max_iter=1000, C=5.0, random_state=SEED),
         X_tfidf_train, X_tfidf_test, y_train, y_test,
         X_tfidf_val, y_val, vectorizer_name='TF-IDF')

best_clf, best_pred = evaluate(
    'TF-IDF + LinearSVC',
    LinearSVC(C=1.0, max_iter=2000, random_state=SEED),
    X_tfidf_train, X_tfidf_test, y_train, y_test,
    X_tfidf_val, y_val, vectorizer_name='TF-IDF'
)

evaluate('TF-IDF + Random Forest',
         RandomForestClassifier(n_estimators=200, random_state=SEED),
         X_tfidf_train, X_tfidf_test, y_train, y_test,
         X_tfidf_val, y_val, vectorizer_name='TF-IDF')

print('\nRapport détaillé (TF-IDF + LinearSVC) :')
print(classification_report(y_test, best_pred, target_names=CLASSES))
````

````python
print('=== Word2Vec (embeddings sémantiques) ===')

evaluate('Word2Vec + Logistic Regression',
         LogisticRegression(max_iter=1000, C=1.0, random_state=SEED),
         X_w2v_train, X_w2v_test, y_train, y_test,
         X_w2v_val, y_val, vectorizer_name='Word2Vec')

evaluate('Word2Vec + LinearSVC',
         LinearSVC(C=1.0, max_iter=2000, random_state=SEED),
         X_w2v_train, X_w2v_test, y_train, y_test,
         X_w2v_val, y_val, vectorizer_name='Word2Vec')

evaluate('Word2Vec + Random Forest',
         RandomForestClassifier(n_estimators=200, random_state=SEED),
         X_w2v_train, X_w2v_test, y_train, y_test,
         X_w2v_val, y_val, vectorizer_name='Word2Vec')
````

````python
print('=== FastText (robuste OOV) ===')

evaluate('FastText + Logistic Regression',
         LogisticRegression(max_iter=1000, C=1.0, random_state=SEED),
         X_ft_train, X_ft_test, y_train, y_test,
         X_ft_val, y_val, vectorizer_name='FastText')

evaluate('FastText + LinearSVC',
         LinearSVC(C=1.0, max_iter=2000, random_state=SEED),
         X_ft_train, X_ft_test, y_train, y_test,
         X_ft_val, y_val, vectorizer_name='FastText')

evaluate('FastText + Random Forest',
         RandomForestClassifier(n_estimators=200, random_state=SEED),
         X_ft_train, X_ft_test, y_train, y_test,
         X_ft_val, y_val, vectorizer_name='FastText')
````

## 🔁 5b. Validation Croisée — Meilleurs Modèles

Pour renforcer la comparaison rigoureuse demandée par le CDC.

````python
print('=== Validation croisée (StratifiedKFold k=5) ===')
print('Sur train+val (85% des données), vectorisation TF-IDF\n')

# Reconstruire X train+val pour la CV
import scipy.sparse as sp
X_tv_s = np.concatenate([X_train_s, X_val_s])
y_tv   = np.concatenate([y_train, y_val])

tfidf_cv = TfidfVectorizer(max_features=8000, ngram_range=(1,2),
                            sublinear_tf=True, min_df=2)
X_tv_tfidf = tfidf_cv.fit_transform(X_tv_s)

cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=SEED)

cv_models = {
    'TF-IDF + LogReg'    : LogisticRegression(max_iter=1000, C=5.0, random_state=SEED),
    'TF-IDF + LinearSVC' : LinearSVC(C=1.0, max_iter=2000, random_state=SEED),
}

cv_results = {}
for name, clf in cv_models.items():
    scores = cross_val_score(clf, X_tv_tfidf, y_tv,
                              cv=cv, scoring='f1_weighted', n_jobs=-1)
    cv_results[name] = scores
    with mlflow.start_run(run_name=f'CV_{name}'):
        mlflow.log_param('vectorizer', 'TF-IDF')
        mlflow.log_param('classifier', clf.__class__.__name__)
        mlflow.log_param('cv_folds', 5)
        mlflow.log_metric('cv_f1_mean', scores.mean())
        mlflow.log_metric('cv_f1_std', scores.std())
    print(f'{name:35s}  CV F1 = {scores.mean():.4f} ± {scores.std():.4f}')
````

## 🌟 6. Bonus — CamemBERT (Sentence Transformers)

````python
try:
    from sentence_transformers import SentenceTransformer
    ST_AVAILABLE = True
except ImportError:
    subprocess.run([sys.executable, '-m', 'pip', 'install', '-q',
                    'sentence-transformers'], check=False)
    try:
        from sentence_transformers import SentenceTransformer
        ST_AVAILABLE = True
    except ImportError:
        ST_AVAILABLE = False
        print('⚠️  sentence-transformers non installé — bonus ignoré')

if ST_AVAILABLE:
    print('Chargement du modèle multilingue (~400 Mo — patience)...')
    sbert = SentenceTransformer('paraphrase-multilingual-mpnet-base-v2')

    # Indices alignés sur le split 70/15/15
    all_indices = np.arange(len(df))
    train_val_idx, test_idx = train_test_split(
        all_indices, test_size=0.15, stratify=y, random_state=SEED
    )
    train_idx, val_idx = train_test_split(
        train_val_idx, test_size=val_ratio,
        stratify=y[train_val_idx], random_state=SEED
    )

    print('Encodage train...')
    X_sbert_train = sbert.encode(df[TEXT_COL].iloc[train_idx].tolist(),
                                  batch_size=32, show_progress_bar=True)
    print('Encodage val...')
    X_sbert_val   = sbert.encode(df[TEXT_COL].iloc[val_idx].tolist(),
                                  batch_size=32, show_progress_bar=True)
    print('Encodage test...')
    X_sbert_test  = sbert.encode(df[TEXT_COL].iloc[test_idx].tolist(),
                                  batch_size=32, show_progress_bar=True)

    y_tr_sb = y[train_idx]
    y_va_sb = y[val_idx]
    y_te_sb = y[test_idx]

    print(f'SBERT → train:{X_sbert_train.shape}  val:{X_sbert_val.shape}  test:{X_sbert_test.shape}')

    print('\n=== CamemBERT / Sentence Transformers ===')
    evaluate('SBERT + Logistic Regression',
             LogisticRegression(max_iter=1000, C=1.0, random_state=SEED),
             X_sbert_train, X_sbert_test, y_tr_sb, y_te_sb,
             X_sbert_val, y_va_sb, vectorizer_name='SBERT')

    evaluate('SBERT + LinearSVC',
             LinearSVC(C=1.0, max_iter=2000, random_state=SEED),
             X_sbert_train, X_sbert_test, y_tr_sb, y_te_sb,
             X_sbert_val, y_va_sb, vectorizer_name='SBERT')
else:
    print('Bonus SBERT ignoré.')
````

## 📊 7. Comparaison Finale

````python
df_res = pd.DataFrame(results).sort_values('F1_Test', ascending=False).reset_index(drop=True)

print('=== RÉSUMÉ COMPARATIF ===')
display(df_res.round(4))

# Graphique comparatif
fig, ax = plt.subplots(figsize=(13, max(5, len(df_res)*0.45)))

colors = plt.cm.RdYlGn(df_res['F1_Test'])
bars = ax.barh(np.arange(len(df_res)), df_res['F1_Test'],
               height=0.5, color=colors, edgecolor='white')

for bar, row in zip(bars, df_res.itertuples()):
    val_str = f'Val={row.F1_Val:.3f}  ' if row.F1_Val is not None else ''
    ax.text(bar.get_width() + 0.005, bar.get_y() + bar.get_height()/2,
            f'{val_str}Test={row.Acc_Test:.3f}', va='center', fontsize=8)

ax.set_yticks(np.arange(len(df_res)))
ax.set_yticklabels(df_res['Modèle'], fontsize=9)
ax.set_xlabel('F1-score (weighted) — Test set')
ax.set_title('Comparaison NLP — Vectorisation × Classificateur')
ax.set_xlim(0, 1.18)
ax.axvline(x=df_res['F1_Test'].max(), color='red',
           linestyle='--', alpha=0.6, label='Meilleur F1')
ax.legend()
ax.grid(axis='x', alpha=0.3)
plt.tight_layout()
plt.savefig('outputs/nlp_comparison.png', dpi=120)
plt.show()

best = df_res.iloc[0]
print(f'\n🏆 Meilleur modèle : {best["Modèle"]}')
print(f'   F1-Test   : {best["F1_Test"]:.4f}')
print(f'   Acc-Test  : {best["Acc_Test"]:.4f}')
````

## 🗂️ 8. Matrice de Confusion — Meilleur Modèle

````python
# TF-IDF + LinearSVC (généralement le meilleur)
final_clf = LinearSVC(C=1.0, max_iter=2000, random_state=SEED)
final_clf.fit(X_tfidf_train, y_train)
y_final_pred = final_clf.predict(X_tfidf_test)

print('Rapport de classification (TF-IDF + LinearSVC) :')
print(classification_report(y_test, y_final_pred, target_names=CLASSES))

plot_confusion(y_test, y_final_pred, CLASSES,
               'Matrice de confusion TF-IDF LinearSVC')
````

## 💾 9. Sauvegarde + Enregistrement MLflow Model Registry

````python
import pickle

# Pipeline final entraîné sur TOUT le corpus (train+val+test)
best_pipeline = Pipeline([
    ('tfidf', TfidfVectorizer(max_features=8000, ngram_range=(1,2),
                               sublinear_tf=True, min_df=2)),
    ('clf',   LinearSVC(C=1.0, max_iter=2000, random_state=SEED))
])
X_all_stem = np.concatenate([X_train_s, X_val_s, X_test_s])
y_all      = np.concatenate([y_train, y_val, y_test])
best_pipeline.fit(X_all_stem, y_all)

# Sauvegarde locale
with open('outputs/nlp_best_pipeline.pkl', 'wb') as f:
    pickle.dump({'pipeline': best_pipeline, 'label_encoder': le}, f)
df_res.to_csv('outputs/nlp_results.csv', index=False)

# ── Enregistrement dans MLflow Model Registry ─────────────────
with mlflow.start_run(run_name='NLP_Best_Pipeline_Final') as run:
    mlflow.log_param('vectorizer', 'TF-IDF_unigram_bigram')
    mlflow.log_param('classifier', 'LinearSVC')
    mlflow.log_param('training_set', 'full_corpus_70+15+15')
    mlflow.log_param('split_ratio', '70/15/15')

    # Métriques du meilleur run
    best_row = df_res[df_res['Modèle'] == 'TF-IDF + LinearSVC'].iloc[0]
    mlflow.log_metric('final_f1_test',  best_row['F1_Test'])
    mlflow.log_metric('final_acc_test', best_row['Acc_Test'])
    if best_row['F1_Val'] is not None:
        mlflow.log_metric('final_f1_val', best_row['F1_Val'])

    # Log artefacts
    mlflow.log_artifact('outputs/nlp_results.csv')
    mlflow.log_artifact('outputs/nlp_comparison.png')

    # Enregistrement du pipeline sklearn
    mlflow.sklearn.log_model(
        best_pipeline,
        artifact_path='nlp_pipeline',
        registered_model_name='EcoSmart_NLP_Pipeline'  # Model Registry
    )

    run_id = run.info.run_id

print('✅ Sauvegardés :')
print('   outputs/nlp_best_pipeline.pkl')
print('   outputs/nlp_results.csv')
print('   outputs/nlp_comparison.png')
print(f'   MLflow run_id : {run_id}')
print('   MLflow Model Registry : EcoSmart_NLP_Pipeline')

# ── Test du pipeline ──────────────────────────────────────────
test_texts = [
    'Plaque en acier inoxydable très résistante à la corrosion',
    'Panneau de bois massif pour cuisine et ameublement',
    'Plastique recyclé haute densité léger et flexible',
]
test_proc   = [preprocess(t, method='stem') for t in test_texts]
predictions = le.inverse_transform(best_pipeline.predict(test_proc))

print('\nTest du pipeline :')
for text, pred in zip(test_texts, predictions):
    print(f'  [{pred:12s}]  {text}')
````
